### CarPole-v0例子（可运行）



在游戏CarPole-v0中，有两个动作，分别是0和1，0表示向左移动小车，1表示向右移动小车。目的是通过移动小车来控制杆的平衡。

我们希望他能维持杆子直立尽可能长的时间。因此每一步都是一个奖励+1，一直到杆子倒下或者小车移出边界为止。

我们希望奖励尽可能多（也就是杆子尽可能长时间直立）。

终止条件：1. 杆子倾斜角度大于±12度； 2. 小车位置超出边界2.4（屏幕边界）； 3. 达到200步上限

### MoutainCar-v0例子

以MoutainCar-v0为例，完整的与gym库交互：
1. 查看环境的动作空间和观测空间
2. 实现一个简单的Agent
3. Agent与环境交互

```python
# 1. 获取环境的动作空间和观测空间
# 更好的是去看官方文档，写的更详细
import gym 
env=gym.make("MountainCar-v0")
print("动作空间={}".format(env.action_space))
print("观察空间={}".format(env.observation_space))
print("观测空间范围={}-{}".format(env.observation_space.low,env.observation_space.high))
print("动作数量={}".format(env.action_space.n))
```

```python
# 2. 实现一个简单的Agent
class SimpleAgent:
    def __init__(self, env):
        pass
    
    def decide(self, observation):
        # 得到小车的位置和速度
        position, velocity = observation
        # 速度下界lb和上界ub，自行设计（通过经验，实验得到）
        lb = min(-0.09*(position+0.25)**2+0.03, 0.3*(position+0.9)**4-0.008)

        ub = -0.07*(position+0.38)**2+0.07

        if lb < velocity < ub:
            action=2  # 向右加速
        else:
            action= 0  # 向左加速
        
        return action

    def learn(self,*args):
        pass

agent=SimpleAgent(env)
```

```python
# 3. 交互函数
def play(env,agent,render=False,train=False):
    episode_reward=0
    observation=env.reset()

    while True:
        if render:
            env.render()
        action = agent.decide(observation)
        next_obervation, reward, done, _ = env.step(action)
        episode_reward+=reward # 回合的累计奖励

        # 判断是否训练Agent
        if train:
            print("================== 训练中... ==================")
            agent.learn(observation,action,reward,next_obervation,done)

        # 如果达到停止条件，则退出循环
        if done:
            print("==========完成=============")
            break

        observation=next_obervation
    return episode_reward # 最后返回奖励
```

```python
# 开始交互一回合

env.seed(3) # 设置随机种子，保证结果可复现
episode_reward=play(env,agent,render=True,train=False)
print("回合奖励：{}".format(episode_reward))

# 如果有渲染界面，就一定要关闭
env.close()
```

```python
# 为了更好的评估性能，我们可以运行多个回合，计算平均奖励
num_episodes=10
episode_rewards = [play(env,agent) for _ in range(num_episodes)]
print("平均回合奖励：{}".format(sum(episode_rewards)/num_episodes))
```
